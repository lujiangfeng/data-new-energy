```python
import pandas as  pd
import numpy as np
import xgboost as xgb
import random
import os 
from datetime import date,datetime
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import roc_curve,roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import StandardScaler as SS
from sklearn.preprocessing import MinMaxScaler as MM
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import LabelBinarizer as LB
from sklearn.externals import joblib
```

```python
data = pd.DataFrame({'pet':      ['cat', 'dog', 'dog', 'fish', 'cat', 'dog', 'cat', 'fish'],
                     'children': [4., 6, 3, 3, 2, 3, 5, 4],
                    'salary':   [90, 24, 44, 27, 32, 59, 36, 27]})
data
```

```python
# 然后创建一个DataFrameMapper ，它是其中最重要的转换器，其中我们在其声明中要配置好，针对哪些列做怎么样的特征转换操作,由于我对 sklearn 的几个转化类都做了 alias ，我们可以之间使用alias
mapper = DataFrameMapper([
   (['pet'], LB()),
   (['children'],SS()),
   (['children',"salary"],MM()) 
 ])
```

```python
# 执行转化操作
np.round(mapper.fit_transform(data.copy()),2)
```

```python
mapper.transformed_names_
```

```python
# 间接生成
dataArray=np.round(mapper.fit_transform(data.copy()),2) 
featureList=mapper.transformed_names_
df4=pd.DataFrame(dataArray  ,columns=featureList)
df4
```

```python
# 直接生成
mapper = DataFrameMapper([
   (['pet'], LB()),
   (['children'],SS()),
   (['children',"salary"],MM()) 
 ],df_out=True)
RealDF=np.round(mapper.fit_transform(data.copy()),2)
RealDF
```
